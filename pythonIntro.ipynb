```python
x = np.linspace(0, 5, 100)
# print(x)
y = np.sin(4*x)*np.exp(-x)
z = np.cos(4*x)*np.exp(-x)

plt.figure(dpi=300)
plt.plot(x, y, 'r--', label = '$e^{-x} sin(4x)$', linewidth = 1.0)
plt.plot(x, z, label = '$e^{-x} cos(4x)$')
plt.xlabel('x')
plt.ylabel('value')
plt.legend()
plt.title('PLOT')
# plt.savefig('figure.png')
```

```python
import pandas as pd  #2 dimansional dataframe with labeled axis
import matplotlib.pyplot as plt

df = pd.read_csv("carprices.csv")  
print(df.head()) # returns the first n rows for the object, quickly testing the object, default n=5

fig, ax = plt.subplots() 

plt.scatter(df['Mileage'],df['Sell Price'])
ax.set_title('Mileage vs Price')
ax.set_xlabel('Mileage')
ax.set_ylabel('Sell Price')
plt.show()

fig, ax =  plt.subplots()
plt.scatter(df['Age'],df['Sell Price'])
ax.set_title('Age vs Price')
ax.set_xlabel('Age')
ax.set_ylabel('Price')
plt.show()

X = df[['Mileage','Age']]
y = df['Sell Price']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)

#linear regression model
from sklearn.linear_model import LinearRegression
clf = LinearRegression() #classifier an object of this class
clf.fit(X_train, y_train)

pred = clf.predict(X_test) #predicted values 
print("predicted values ", )


for i in pred:   #vertical print
    print(round(i))

Accuracy = clf.score(X_test, y_test) #accuracy of model
print("Accuracy = ", Accuracy)
print("Accuracy = ", round(Accuracy,2)*100 , "%") #strogulopoihmeno
```
