```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = np.linspace(0, 5, 100)
# print(x)
y = np.sin(4*x)*np.exp(-x)
z = np.cos(4*x)*np.exp(-x)

plt.figure()
plt.plot(x, y, 'r--', label = '$e^{-x} sin(4x)$', linewidth = 4.0)
plt.plot(x, z, label = '$e^{-x} cos(4x)$')
plt.xlabel('x')
plt.ylabel('value')
plt.legend()
plt.title('PLOT')
# plt.savefig('figure.png')
```
