```python
# Simple python list, something like an array, with 20 elements
# and initialized with ones
pa = [1]*20

print('This is the first array:')
print(pa)

# This is what it takes to multiply every element by 3
# Using a simple for loop
for i in range(len(pa)):
    pa[i] *= 3

print('This is it after the multiplication:')
print(pa)
```

```python
# Here comes NUMPY to save the day, lets import it!
import numpy as np

# Now let's make a simple array of 20 elements initialized with ones as before
a = np.ones(20)
print(a)

# Let's do the same as before and multiply every element
a *= 3
print(a)
# Easier, right?

# Now lets make an array with 5 random numbers from 0 to 10
a = np.random.randint(0, 10, 5)
print(a)

# This is a two dimentional array with 10 rows and 2 columns
a = np.ones((10, 2))
print(a)

# A linearly spaced array of 20 elements from 0 to 10
a = np.linspace(0, 10, 20)
print(a)

# Various other functions
a.sum()
a.min()
a.max()
a.mean()
a.var()
a.std()

# This chooses the element in the index 0 to the element in the index 5 - 1 = 4
# Don't forget that arrays start at 0!
a[:5]

# Various other useful features
a.size
a.shape
type(a)
a.dtype
a
```

```python
# Now we will plot some of that stuff using the 
# Matplotlib library! Let's import it!
import matplotlib.pyplot as plt

# This will be our x axis values that we will feed into the sine function!
# We use numpy to do this, using 0.1 as a step, from 0 to 3Ï€
x = np.arange(0, 3 * np.pi, 0.1)

# This is the function, also using numpy!
y = np.sin(x) 

# Let's plot it!
plt.title("sine wave form") 
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
# Another simple example!
# If this does not work, do not forget to run the previous cells above
# or include the libraries here! (numpy and matplotlib)

x = np.linspace(0, 5, 100)
y = np.sin(4*x)*np.exp(-x)
z = np.cos(4*x)*np.exp(-x)

plt.figure(dpi=200)
plt.plot(x, y, 'r--', label = '$e^{-x} sin(4x)$', linewidth = 1.0)
plt.plot(x, z, label = '$e^{-x} cos(4x)$')
plt.xlabel('x')
plt.ylabel('value')
plt.legend()
plt.title('PLOT')
plt.show()
# plt.savefig('figure.png')
```

```python
import pandas as pd  #2 dimansional dataframe with labeled axis
import matplotlib.pyplot as plt

df = pd.read_csv("https://bit.ly/2YrmGvw")  
print(df.head()) # returns the first n rows for the object, quickly testing the object, default n=5

plt.scatter(df['Mileage'],df['Sell Price'])
plt.title('Mileage vs Price')
plt.xlabel('Mileage')
plt.ylabel('Sell Price')
plt.show()

plt.scatter(df['Age'], df['Sell Price'])
plt.title('Age vs Price')
plt.xlabel('Age')
plt.ylabel('Price')
plt.show()

X = df[['Mileage','Age']]
y = df['Sell Price']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3)

#linear regression model
from sklearn.linear_model import LinearRegression
clf = LinearRegression() #classifier an object of this class
clf.fit(X_train, y_train)

pred = clf.predict(X_test) #predicted values 
print("predicted values ", )


for i in pred:   #vertical print
    print(round(i))

Accuracy = clf.score(X_test, y_test) #accuracy of model
print("Accuracy = ", Accuracy)
print("Accuracy = ", round(Accuracy,2)*100 , "%") #strogulopoihmeno
```
